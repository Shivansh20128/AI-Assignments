```python
import pandas as pd
import bnlearn

df = pd.concat([pd.DataFrame(X, columns=wine.data.feature_names), pd.Series(y, name='target')], axis=1)

# Learn the structure of the Bayesian network
structure = bnlearn.structure_learning.fit(df, methodtype='hc', scoretype='bic')

# Visualize the Bayesian network structure
bnlearn.plot(structure)

# Fit the parameters of the Bayesian network
model = bnlearn.structure_learning.fit(df, methodtype='hc', scoretype='bic')

# Plot the probability distribution
bnlearn.plot(model)
```
